```python
droplist = [
    'radius_mean', 'texture_mean', 'perimeter_mean', 'compactness_mean', 'concavity_mean', 'fractal_dimension_mean', 
    'perimeter_se', 'area_se', 'compactness_se', 'concavity_se', 'radius_worst', 'smoothness_worst', 'compactness_worst', 'concave points_worst'
    ]

x_feature_select = x_scaled.drop(droplist, axis=1)
print(x_feature_select.columns)
```

```python
f, ax = plt.subplots(figsize=(14, 14))
sns.heatmap(x_feature_select.corr(), annot=True, linewidths=.5, fmt='.1f', ax=ax)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.metrics import accuracy_score
import xgboost 
from sklearn.metrics import classification_report 
from sklearn import metrics 
from sklearn.metrics import recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score, cross_val_predict
```

### K-Nearest Neighbor 

```python
x_train, x_test, y_train, y_test = train_test_split(x_feature_select, y, test_size=0.28, random_state=42)
for i in range(1, 17):
    neigh = KNeighborsClassifier(n_neighbors=i)
    cv = RepeatedKFold(n_splits=10, n_repeats=10, random_state=42)
    scores=cross_val_score(neigh, x_test, y_test, scoring='accuracy', cv=cv)
    print('k = %d : %.3f' % (i, np.mean(scores)))
```

```python
y_pred = cross_val_predict(neigh, x_test, y_test, cv=10)
cf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(cf_matrix, annot=True, fmt='.0f', cmap='PuRd')
plt.show()
```

### Random Forest

```python
x_train, x_test, y_train, y_test = train_test_split(x_feature_select, y, test_size=0.28, random_state=42)
clf_rf = RandomForestClassifier(random_state=42)
clr_rf = clf_rf.fit(x_train, y_train)
clr = clr_rf.predict(x_test)
ac = accuracy_score(y_test, clr)*100
print('Accuracy is: ', ac)
print(classification_report(y_test, clr, target_names=['Malignant (Class 0)', 'Benign (Class 1)']))
cm = confusion_matrix(y_test, clf_rf.predict(x_test))
sns.heatmap(cm, annot=True, fmt='d')
```

### Logistic Regression 

```python
x_train, x_test, y_train, y_test = train_test_split(x_feature_select, y, test_size=0.28, random_state=42)
classifier_lr = LogisticRegression()
classifier_lr.fit(x_train, y_train)

prediction_lr = classifier_lr.predict(x_test)
print(classification_report(y_test, prediction_lr, target_names=['Malignant (Class 0)', 'Benign (Class 1)']))
```

```python
print( 'Accuracy of Logistic Regression model is {}'.format(accuracy_score(y_test, prediction_lr)*100) )
```

```python
cm = confusion_matrix(y_test, classifier_lr.predict(x_test))
sns.heatmap(cm, annot=True, fmt='d')
```

### XG Boost

```python
x_train, x_test, y_train, y_test = train_test_split(x_feature_select, y, test_size=0.28, random_state=42)
xgb = xgboost.XGBClassifier(earning_rate=0.009, n_estimators=350, subsample=0.8, max_depth=7)
xgb.fit(x_train, y_train)
prediction_xgb = xgb.predict(x_test)
print(classification_report(y_test, prediction_xgb, target_names=['Malignant (Class 0)', 'Benign (Class 1)']))
```

```python
print('Accuracy of XGB model is {}'.format(accuracy_score(y_test, prediction_xgb)*10))
```

```python
cm = confusion_matrix(y_test, xgb.predict(x_test))
sns.heatmap(cm, annot=True, fmt='d')
```

### SVM

```python
x_train, x_test, y_train, y_test = train_test_split(x_feature_select, y, test_size=0.28, random_state=42)
clfSvm = svm.SVC()
clfSvm.fit(x_train, y_train)
prediction_svm = clfSvm.predict(x_test)
print(classification_report(y_test, prediction_svm, target_names=['Malignant (Class 0)', 'Benign (Class 1)']))
```

```python
print('Accuracy of XGB model is {}'.format(accuracy_score(y_test, prediction_svm)*10))
```

```python
cm = confusion_matrix(y_test, clfSvm.predict(x_test))
sns.heatmap(cm, annot=True, fmt='d')
```
